```python
test_data = pd.read_csv('test.csv')
```

```python
# X0 thru X8 are categorical

for i in range(9):
    k = i + 1 # start at column 'X0'
    df.iloc[:, k] = df.iloc[:, k].astype("category").cat.codes
    
# df.dtypes
```

```python
train, test = train_test_split(df, test_size = 0.2)

# train sets
train_x = train.drop('y', 1)
train_y = train['y']

# test sets
test_x = test.drop('y', 1)
test_y = test['y']
```

## Random Forest w. test/train

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# model
forest = RandomForestRegressor()

# train
forest.fit(train_x, train_y)

# predict
forest_predictions = forest.predict(test_x)

# feature importances
features = pd.DataFrame({'Importance': forest.feature_importances_}, index = train_x.columns).sort_values('Importance', ascending = False)

# confusion matrix
pd.crosstab(test_y, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
```

```python
features
```

```python
new = features.loc[features['Importance'] > 0.002]
features_necessary = []
for i in new.index:
    features_necessary.append(i)
features_necessary
```
