# Facebook Challenge

```python
import numpy as np
import pandas as pd

# read in data
bids = pd.read_csv('bids.csv')
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')


# print(bids.shape)
# print(bids.head())
# print(train_set.shape)
# print(test_set.shape)
```

# Merging Data

```python
# Merge using bidder_id for TRAIN_SET
bid_total = bids.merge(train_set, on='bidder_id')
bid_total = bid_total.set_index('bid_id', drop=True)
# print(bid_total.head())

# Merge using bidder_id for TEST_SET
test_total = bids.merge(test_set, on='bidder_id')
test_total = test_total.set_index('bid_id', drop=True)
# print(test_total.head())
```

# Cleaning NaNs

```python
# TRAIN DATA 
bid_total['payment_account'].fillna(0, inplace=True)
bid_total['address'].fillna(0, inplace=True)
bid_total['outcome'].fillna(0, inplace=True)
# print(bid_total.head())

# TEST DATA - every inf is a NaN now
pd.options.mode.use_inf_as_na = True

# TEST DATA - delete every row containing a NaN
test_total = test_total.dropna(axis=0, how='any', inplace=True)

# Check values for NaN or null, infinity or a value
print(test_total.isnull().values.any())
```

# Normalize Time

```python
mean_time = bid_total.time.mean()
max_time = bid_total.time.max()
min_time = bid_total.time.min()
bid_total['time'] = bid_total['time'].apply(lambda x: ( x - min_time ) / ( mean_time ))
```

```python
mean_time = test_total.time.mean()
max_time = test_total.time.max()
min_time = test_total.time.min()
test_total['time'] = test_total['time'].apply(lambda x: ( x - min_time ) / ( mean_time ))
```

# Splitting Data

```python
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier


train, test = train_test_split(bid_total, test_size = 0.2)

# train sets
train_x = train.drop('outcome', 1)
train_y = train['outcome']

# test sets
test_x = test.drop('outcome', 1)
test_y = test['outcome']
```

# Random Forest

```python
# Maybe random forest is not possible?
# # model
tree = DecisionTreeClassifier(criterion = "entropy")

# # train
tree.fit(train_x, train_y)

# # predict
tree_predictions = tree.predict(test_x)

# # I put the attribute names and their respective information gains in a data frame for readability.
pd.DataFrame({'Gain': tree.feature_importances_}, index = train_x.columns).sort_values('Gain', ascending = False)
```

# Confusion Matrix

```python
pd.crosstab(test_y, tree_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(test_y, tree_predictions, normalize=True, sample_weight=None)
accuracy
```

```python
# # Testing predictions
testing_predictions = tree.predict(test_total)
pd.crosstab(test_y, tree_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
accuracy = metrics.accuracy_score(test_y, testing_predictions, normalize=True, sample_weight=None)
accuracy
```
