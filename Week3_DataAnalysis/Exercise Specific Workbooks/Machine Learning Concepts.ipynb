```python
# CARTESIAN SOLUTION --------------------------------------------------------------------------------------------------
# Q. Use a graphical approach to finding a threshold on x and y to differentiate between red/blue points.
#    - You should only do this with the train partition, right?

from matplotlib import pyplot as plt

red = known.loc[known['colour']== 'red']
blue = known.loc[known['colour']== 'blue']

plt.scatter(red['x'],red['y'], c='r')
plt.scatter(blue['x'],blue['y'], c='b')
```

```python
# Q. Write a function which will use x and y coordinates to differentiate between red/blue points.
def func(x,y):
    if x<1 and x>-1 and y<1 and y>-1:
        return 'red'
    return 'blue'
```

```python
# Q. Use this function to make predictions for the test partition.
#    - Calculate accuracy.
#    - Generate a confusion matrix. 

predictions = []
for i in range(len(known)):
    x = known.iloc[i,0]
    y = known.iloc[i,1]
    # x is the 0 column
    predictions.append(func(x,y))
    
predictions = pd.Series(predictions)
```

```python
# ENGINEER FEATURES ---------------------------------------------------------------------------------------------------

# Q. Add a new feature, r = sqrt(x**2 + y**2), to the data.
pd.crosstab(known['colour'], predictions, rownames=['Actual'], colnames = ['Predicted'], margins = True)
```

```python
# POLAR SOLUTION ------------------------------------------------------------------------------------------------------

# Q. Use a graphical approach to finding a threshold on r.
# Q. Write a function which will use r to differentiate between red/blue points.
# Q. Use this function to make predictions for the test partition.
#    - Calculate accuracy.
#    - Generate a confusion matrix.

def foo(r):
    if r < 1:
        return 'red'
    return 'blue'

predictions = []
for i in range(len(known)):
    r = known.iloc[i,3]
    # x is the 0 column
    predictions.append(foo(r))
    
predictions = pd.Series(predictions)

pd.crosstab(known['colour'], predictions, rownames=['Actual'], colnames = ['Predicted'], margins = True)
```
