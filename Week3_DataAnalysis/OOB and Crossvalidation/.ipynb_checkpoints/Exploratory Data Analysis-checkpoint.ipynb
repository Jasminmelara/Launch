# Exploratory Data Analysis

```python
import pandas as pd 
import numpy as np 
import scipy as sp 
from sklearn import datasets

# Load the star catalog data.

starcat = pd.read_csv("starcat.csv")
starcat.head()
# Define a factor variable.
#
# Q. Define a "nakedeye", categorical variable which indicates whether or not a star is visible to the naked eye. You
#    might find https://en.wikipedia.org/wiki/Apparent_magnitude useful.

# If Vmag < 7, then visible

starcat["nakedeye"] = 1

# if starcat["Vmag"] > 7:
#     nakedeye.append(loc.starcat["Vmag"])

for i in range(len(starcat)):
    if starcat.loc[i,"Vmag"] > 7:
        starcat.loc[i,"nakedeye"] = 0
# print(starcat)


# Q. Create an overview of the data using describe().
starcat.describe()

# Q. By looking at the range for rrr (Declination) and resRA (Right Ascension) can we infer their meaning? No
# Q. Define a "hemisphere" categorical variable based on rrr. 90 to -90. 
# Northern Hemi 0 to 90 {hemisphere = 0} ; Southern Hemi -90 to 90 {hemisphere = 1}

starcat["hemisphere"] = 1

for i in range(len(starcat)):
    if starcat.loc[i, "rrr"] > 0:
        starcat.loc[i, "hemisphere"] = 0
# print(starcat)
        
# Q. Find the star which is closest to the Celestial Equator.
# Step 1: Take the absolute value of rrr
starcat["rrr"] = starcat['rrr'].abs()
# Step 2: Organize is ascending order
ordered = starcat.sort_values(by="rrr", ascending = "True")
# Step 3: Take the head(1)

print(ordered)
```

## Exploring Islands 

```python
# EXPLORING ISLANDS ---------------------------------------------------------------------------------------------------

# The islands data gives the areas of major land masses in square miles.
#

islands = pd.read_csv("islands.csv")
islands.head(10)
# Q. Convert the data from square miles to square kilometres.
# Conversion Factor - mi^2 to km^2
factor = 1.609344 * 1.609344
islands.x = islands.x * factor


# Q. How many observations are there?
print((len(islands)))

# Q. Find the mean and median area. Discuss.
print("Mean:", islands.mean())
print("Median:", islands.median())

# Mean: 3244 --- Median: 106
# There are a bunch of little islands that have visible stars

# Q. Find the area of the largest and smallest land masses. There are multiple ways to do this.
print("Max", islands.max())
print("Min", islands.min())

# Max is Victoria with 43998 km^2, Min is Africa with 31 km^2

# Q. Find the quartiles of area.

# So this is a bit of a brute method
# print("Quantile 1", islands.quantile(.25))
# print("Quantile2", islands.quantile(.5))
# print("Quantile3", islands.quantile(.75))

# print(islands.columns)
# THis is solid for 50% but what about the remaining?
# print("The 50% quantile of island area is:" , islands["x"].quantile())
print(pd.qcut(islands.x,4))
      
# Q. Create a histogram of area. Consider how you could make this more informative.
# islands.hist()
```
